```python
from pathlib import Path
import json
import folium

cwd = Path.cwd()
target = cwd.parent / "data" / "world_cities_by_time-zone.json"

with open(target, "r") as f:
    cities_by_zone = json.load(f)

zone_cities = cities_by_zone["Europe/Andorra"]
print(f"Number of cities in zone: {len(zone_cities)}")
```

```python
msu_lat = 33.91357
msu_lon = -98.51947

m = folium.Map(location=[msu_lat, msu_lon], zoom_start=2)

folium.Marker([msu_lat, msu_lon], popup="MSU", icon=folium.Icon(color="red")).add_to(m)

for city in zone_cities:
    city_lat = float(city["lat"])
    city_lon = float(city["lon"])

    folium.Marker(
        [city_lat, city_lon],
        popup=city["city-name"]
    ).add_to(m)

    folium.PolyLine(
        locations=[[msu_lat, msu_lon], [city_lat, city_lon]],
        color="blue",
        weight=2,
        opacity=0.6
    ).add_to(m)
```
