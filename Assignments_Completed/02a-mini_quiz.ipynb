### 3️⃣ Quiz Question — Changing Tuple Elements

Given the following tuple:

```python
T = (13, 17, 31, 37, 73)
```

Write the necessary Python code that would change the second element of the tuple to 11.

````python
# Answer 3
T = (13,17,31,37,73)

T_list = list(T)
T_list[1] = 11
T = tuple(T_list)

T
````
